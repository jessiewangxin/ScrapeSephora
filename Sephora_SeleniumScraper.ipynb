```python
from selenium import webdriver
import time 
import random
import csv
import pandas as pd

driver = webdriver.Chrome()

#used to store data 
dict_data={'category':[],'brand':[],'prod_name':[], 'num_loves':[], 'num_prodsizes':[], \
           'price':[],'skintype':[], 'skinconcern':[],'ingredients':[], \
          'num_five':[],'num_four':[],'num_three':[],'num_two':[],'num_one':[] }

#skincare categories we want to scrape 
'''url_categories =['eye-treatment-dark-circle-treatment', 'vegan-skin-care','clean-skin-care', \
                 'moisturizing-cream-oils-mists', 'moisturizing-cream-oils-mists', 'cleanser', \
                 'facial-treatments', 'wellness-skincare', 'eye-treatment-dark-circle-treatment',\
                 'sunscreen-sun-protection', 'self-tanning-products', 'lip-treatments',\
                 'skin-care-sets-travel-value','mini-skincare']'''

url_categories =['clean-skin-care']


#going through individual categories listed above
for url in url_categories:
    driver.get(f"https://www.sephora.com/shop/{url}")
    
    #slowly scrolls to the end of the page to get number of pages   
    i = 0
    while (i < 5):
       #print(i)
        start = 1000*i
        end = 1000*(i+1)
        driver.execute_script(f"window.scrollTo({start},{end})")
        time.sleep(1)
        i += 1
        
    #number of pages within each skincare category
    pagelinks = driver.find_elements_by_class_name("css-17n3p1l")
    maxpages = int(pagelinks[len(pagelinks)-1].text)
        #print(maxpages)
        #maxpages = 2

    n=0    
    while n <maxpages:
        #print(i)
        time.sleep(3)
        driver.get(f'https://www.sephora.com/shop/{url}?currentPage={n+1}')
        n+=1

        #this collects all the product links on each page  
        i = 0
        while (i < 5):
           #print(i)
            start = 1000*i
            end = 1000*(i+1)
            driver.execute_script(f"window.scrollTo({start},{end})")
            time.sleep(1)
            i += 1

        #list of product links 
        
        #key page #
        #value information
        #dictionary can be converted to df 
        
        links = [link.get_attribute('href') for link in driver.find_elements_by_xpath('//div[@class="css-12egk0t"]/a')]
        testlinks = links #[:2]
        #this clicks on each product link 
        for link in testlinks:

            driver.get(link)

            category=url 
            brand = driver.find_elements_by_class_name("css-euydo4")[0].text
            prodnameraw = driver.find_elements_by_class_name("css-0")
            prod_name = [i for i in prodnameraw if i.text!=''][0].text
            num_loves = driver.find_elements_by_class_name("css-2rg6q7")[1].text 
            num_prodsizes = len(driver.find_elements_by_class_name("css-6dchx0"))
            #standardsize = driver.find_elements_by_class_name("css-6dchx0")[0].text
            price = driver.find_element_by_xpath('/html/body/div[3]/div[5]/main/div[2]/div[1]/div/div/div[2]/div[1]/div[2]/div[1]/span').text

            driver.find_element_by_xpath('//*[@id="tab0"]/span').click()
            product_detail=driver.find_elements_by_class_name("css-pz80c5 ")
            detail_list=[i for i in product_detail if i.text!=''][0].text
            detail_list=detail_list.split('\n')
            skintype = [i for i in detail_list if i.startswith('Skin Type:')]
            skinconcern = [i for i in detail_list if i.startswith('Skincare Concerns:')]

            driver.find_element_by_xpath('//*[@id="tab2"]/span').click()
            ingredientspage = driver.find_elements_by_class_name("css-pz80c5 ")
            ingredientspage_trim=[i for i in ingredientspage if i.text!='']
            if len(ingredientspage_trim)!= 0:
                ingredientspage_trim=ingredientspage_trim[0].text.split('\n')
            else:
                ingredientspage_trim=['blank']
                
            #JW NOTE: NEED TO FIX THIS
            ingredients = ''
            i=0

            while i<len(ingredientspage_trim)+1 and ingredients=='':
                text = ingredientspage_trim[i]
                #print(i)
                if "Water, " in text:
                    ingredients = text
                if "Water , " in text:
                    ingredients = text
                elif "Oil, " in text:
                    ingredients = text
                elif "Oil , " in text:
                    ingredients = text
                elif "Acid, " in text:
                    ingredients = text
                elif "Acid , " in text:
                    ingredients = text
                elif "Alcohol, " in text:
                    ingredients = text
                elif "Alcohol , " in text:
                    ingredients = text
                elif i == len(ingredientspage_trim)-1:
                    ingredients = 'not found'
                i+=1
        
            i = 0
            while (i < 5):
               #print(i)
                start = 500*i
                end = 500*(i+1)
                driver.execute_script(f"window.scrollTo({start},{end})")
                time.sleep(1)
                i += 1
            driver.execute_script(f"window.scrollTo({end},document.body.scrollHeight)")
            
            rating_raw = driver.find_elements_by_class_name("css-1r36mik ")
            #rating_raw2 = [i for i in rating_raw if i.text!=''][0].text
            
            #num_reviews_raw =  driver.find_elements_by_class_name("css-tc6qfq ")
            #num_reviews= [i for i in num_reviews_raw if i.text!=''][0].text
            if driver.find_elements_by_class_name("css-2rg6q7")[0].text=='0 reviews':
                num_five = '0'
                num_four = '0'
                num_three = '0'
                num_two = '0'
                num_one = '0'
            elif driver.find_elements_by_class_name("css-bclwec ")!=[]:
                starsummary = driver.find_elements_by_class_name("css-bclwec ")[0].text.split('\n')
                num_five = starsummary[1]
                num_four = starsummary[3]
                num_three = starsummary[5]
                num_two = starsummary[7]
                num_one = starsummary[9]
            else:
                num_five = 0
                num_four = 0
                num_three = 0
                num_two = 0
                num_one = 0
            
            #appending to our data dictionary
            
            dict_data['category'].append(category)
            dict_data['brand'].append(brand)
            dict_data['prod_name'].append(prod_name)
            dict_data['num_loves'].append(num_loves)
            dict_data['num_prodsizes'].append(num_prodsizes)   
            dict_data['price'].append(price) 
            dict_data['skintype'].append(skintype)      
            dict_data['skinconcern'].append(skinconcern) 
            dict_data['ingredients'].append(ingredients)
            #dict_data['rating'].append(rating)
            #dict_data['num_reviews'].append(num_reviews)
            dict_data['num_five'].append(num_five)
            dict_data['num_four'].append(num_four)
            dict_data['num_three'].append(num_three)
            dict_data['num_two'].append(num_two)
            dict_data['num_one'].append(num_one)
             
            #random sleep of 1 to 3 seconds before going to the next link
            x = random.randint(1,3)
            time.sleep(x)
            
            
        
df = pd.DataFrame(dict_data)
df.to_csv('output.csv')
        
```
