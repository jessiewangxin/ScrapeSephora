```python
from selenium import webdriver
import time 
import random
import csv
import pandas 

driver = webdriver.Chrome()

#used to store data 
dict_data={'category':[],'brand':[],'details':[],'ingredients':[],'rating':[],'num_reviews':[]}

#skincare categories we want to scrape 
url_categories =['eye-treatment-dark-circle-treatment']
    #'vegan-skin-care']
                 #,'clean-skin-care', 'moisturizing-cream-oils-mists', \
#                 'moisturizing-cream-oils-mists', 'cleanser', 'facial-treatments', 'wellness-skincare', \
#                 'eye-treatment-dark-circle-treatment', 'sunscreen-sun-protection', \
#                 'self-tanning-products', 'lip-treatments','skin-care-sets-travel-value','mini-skincare']



#going through individual categories listed above
for url in url_categories:
    driver.get(f"https://www.sephora.com/shop/{url}")
    
    #slowly scrolls to the end of the page to get number of pages   
    i = 0
    while (i < 5):
       #print(i)
        start = 1000*i
        end = 1000*(i+1)
        driver.execute_script(f"window.scrollTo({start},{end})")
        time.sleep(1)
        i += 1
        
    #number of pages within each skincare category
    pagelinks = driver.find_elements_by_class_name("css-17n3p1l")
    for test in pagelinks[len(pagelinks)-1:]:
        #maxpages = int(test.text)
        #print(maxpages)
        maxpages = 2
  
    for i in range(0,maxpages):
        time.sleep(3)
        driver.get(f'https://www.sephora.com/shop/{url}?currentPage={i+1}')
    n=0
    while n <4:
        #print(i)
        time.sleep(3)
        driver.get(f'https://www.sephora.com/shop/{url}?currentPage={n+1}')

        #this collects all the product links on each page  
        i = 0
        while (i < 5):
           #print(i)
            start = 1000*i
            end = 1000*(i+1)
            driver.execute_script(f"window.scrollTo({start},{end})")
            time.sleep(1)
            i += 1

        #list of product links 
        
        #key page #
        #value information
        #dictionary can be converted to df 
        
        links = [link.get_attribute('href') for link in driver.find_elements_by_xpath('//div[@class="css-12egk0t"]/a')]
        testlinks = links [:2]
        #this clicks on each product link 
        for link in testlinks:

            driver.get(link)

            category=url 
            brand = driver.find_elements_by_class_name("css-euydo4")[0].text

            driver.find_element_by_xpath('//*[@id="tab2"]/span').click()
            details = driver.find_elements_by_class_name("css-pz80c5 ")

            driver.find_element_by_xpath('//*[@id="tab2"]/span').click()
            ingredients = driver.find_elements_by_class_name("css-pz80c5 ")
            
            i = 0
            while (i < 5):
               #print(i)
                start = 500*i
                end = 500*(i+1)
                driver.execute_script(f"window.scrollTo({start},{end})")
                time.sleep(1)
                i += 1
            
            rating = driver.find_elements_by_class_name("css-1r36mik ")
            num_reviews =  driver.find_elements_by_class_name("css-tc6qfq ")
            
            #appending to our data dictionary
            
            dict_data['category'].append(category)
            dict_data['brand'].append(brand)
            dict_data['details'].append(details)
            dict_data['ingredients'].append(ingredients)
            dict_data['rating'].append(rating)
            dict_data['num_reviews'].append(num_reviews)
            
            #random sleep of 1 to 3 seconds before going to the next link
            x = random.randint(1,4)
            time.sleep(x)

        n=n+1
        
pd.DataFrame(dict_data)
        
        
```
